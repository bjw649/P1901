```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime as dt
#In order to show plots in Jupyter notebook use:
%matplotlib inline
plt.style.use("ggplot")
```

```python
# Read the temp and Anabaena CSV file
fp = "G:\\Team Drives\\FSA Team Drive\\Projects\\1901\\Data\\TempData\\Read_CSV\\CombineData.csv"
df = pd.read_csv(fp)
df["Time"] = pd.to_datetime(df["Time"], errors='coerce', dayfirst=True )
df.head()
```

```python
df.drop(columns=['UTSR6_TG_0.5_7.0','UTSR12_TG_0.5_8.5','UTSR2_TG_0.5_4.5','XYLR6_TG_2.0_7.0'],inplace=True)
#df.drop(columns=["XYLR6_TG_2.0_7.0"],inplace=True)
df.head()
```

```python
df["AnaR2_log10"] = np.log10(df["AnaR2"])
df["AnaR6_log10"] = np.log10(df["AnaR6"])
df["AnaR12_log10"] = np.log10(df["AnaR12"])

df.head()
```

```python
len(df)
```

```python
fig = plt.figure(figsize=(20, 10))


plt.subplot(2,1,1)

plt.scatter(df['Time'].tolist(), df['AnaR2_log10'], c='b', label='Anabaena Cell Counts at R2')
plt.scatter(df['Time'].tolist(), df['AnaR6_log10'], c='r', label='Anabaena Cell Counts at R6')
plt.scatter(df['Time'].tolist(), df['AnaR12_log10'], c='k', label='Anabaena Cell Counts at R12')

plt.xlim([dt.date(2009, 1, 1), dt.date(2019, 1, 1)])


plt.legend(loc='upper left')
plt.ylabel("Cell Counts log10")
plt.title("Cell Counts at R2, R6 and R12 (log10)" )


plt.subplot(2,1,2)
# What is the location of termistor??
plt.plot(df['Time'].tolist(), df['Ther_TG_1.0_8.0'], c= 'g', label='Thermistor')
plt.plot(df['Time'].tolist(), df['UTSR2_TG_0.5_5.0'], c='b', label='UTS R2')
plt.plot(df['Time'].tolist(), df['XYLR2_TG_0.5_10.0'], c='b', label='Xylem R2')

plt.plot(df['Time'].tolist(), df['UTSR6_TG_0.5_7.5'], c='r', label='UTS R6')
plt.plot(df['Time'].tolist(), df['XYLR6_TG_1.0_7.0'], c='r', label='Xylem R6')

plt.plot(df['Time'].tolist(), df['UTSR12_TG_0.5_9.0'], c='k', label='UTS R12')

plt.xlim([dt.date(2009, 1, 1), dt.date(2019, 1, 1)])
#plt.ylim(0,0.6)
plt.title("Temperature Gradients at Sites" )

plt.legend(loc='upper left')
plt.ylabel("Deg C / metre depth")


#plt.xlabel("Time")
plt.savefig("AllTempAndAnabaenaAllYears", dpi=300, bbox_inches='tight')
plt.show()

```
